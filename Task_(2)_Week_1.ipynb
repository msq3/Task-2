## 1. Exploratory Data Analysis (EDA)

### Perform summary statistics on the dataset.

```python
print(df.describe())
```

### Identify and analyze patterns in the data.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.scatterplot(x="Datetime", y="TrafficIndexLive", data=df)
plt.title('Relationship between Date and Traffic Index')
plt.xlabel('Datetime')
plt.ylabel('TrafficIndexLive')
plt.show()
```

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
sns.countplot(x="Datetime", data=df)
plt.title('Distribution of Dates')
plt.xlabel('TrafficIndexLive	')
plt.ylabel('Datetime')
plt.show()
```

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


```python
plt.scatter(df[TrafficIndexLive'], df['Accident_severity'])
plt.xlabel('TrafficIndexLive')
plt.ylabel('Accident_severity')
plt.title('TrafficIndexLive vs Accident_severity')
plt.show()
```
